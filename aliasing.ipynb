```python
import numpy as np
import matplotlib.pyplot as plt

def continuous(split = 1000, start = 0, end = 0.5):

    return np.arange(start, end, 0.001)

def sine_wave_generator(time, frequency):
    sine_wave = np.sin(2 * np.pi * frequency * time)
    return sine_wave

def sampler(start, end, frequency, sample_rate):
    
    new_time = np.arange(start, end, 1 / sample_rate)
    return new_time, np.sin(2 * np.pi * frequency * new_time)




time = continuous()
sine_wave_value = sine_wave_generator(time, 10)
plt.figure(figsize=(20, 10))

plt.subplot(2, 2, 1)
plt.plot(time, sine_wave_value,)
plt.title('First Signal with 10 Hz')
plt.xlabel('time')
plt.ylabel('Amplititude')
plt.axhline()
plt.grid(True)

new_time, sample_value = sampler(0, 0.5, 10, 40)
plt.subplot(2, 2, 2)
plt.stem(new_time, sample_value, linefmt='r-', use_line_collection=True)
plt.title('First signal sampling (40 Hz)')
plt.xlabel('n')
plt.ylabel('Amplititude')
plt.grid(True)

plt.subplot(2, 2, 3)
plt.plot(new_time, sample_value)
plt.stem(new_time, sample_value, linefmt='r-', use_line_collection=True)
plt.title('Reconstructed first signal')
plt.axhline()
plt.xlabel('t(time)')
plt.ylabel('Amplititude')
    
#plt.tight_layout()
plt.show()


#second-singal with 50Hz
sine_wave_value = sine_wave_generator(time, 50)
plt.figure(figsize=(20, 10))
plt.subplot(2, 2, 1)
plt.plot(time, sine_wave_value)
plt.title('Second signal with 50 Hz')
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.axhline()


new_time, sample_value = sampler(0, 0.5, 50, 40)
plt.subplot(2, 2, 2)
plt.stem(new_time, sample_value, linefmt='r-', use_line_collection=True)
plt.title('Second signal sampling (40 Hz)')
plt.xlabel('n')
plt.ylabel('Amplititude')
plt.grid(True)


plt.subplot(2, 2, 3)
plt.stem(new_time, sample_value, linefmt='r-', use_line_collection=True)
plt.plot(new_time, sample_value)
plt.title('Reconstruct second signal')
plt.xlabel('Amplititude')
plt.ylabel('time')
plt.axhline()
plt.grid(True)
plt.show()

```
