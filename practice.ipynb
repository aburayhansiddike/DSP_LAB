```python
#create basic impulse signals
import numpy as np
import matplotlib.pyplot as plt


def unit_ramp_signal_generator(start, end):
    arr = np.arange(start, end + 1)
    y_axis_values = np.empty(len(arr))
    for i in range(len(arr)):
        if arr[i] <= 0:
            y_axis_values[i] = 0
        else:
            y_axis_values[i] = i + start
    
    print(y_axis_values)
    return y_axis_values


def unit_step_signal_generator(start, end):
    arr = np.arange(start, end + 1)
    y_axis_values = np.empty(len(arr))
    for i in range(len(arr)):
        if arr[i] < 0:
            y_axis_values[i] = 0
        else:
            y_axis_values[i] = 1
    
    return y_axis_values

def unit_sample_signal_generator(length, start, end):
    arr = np.arange(start, end + 1)
    y_axis_values = np.empty(len(arr))
    for i in range(len(arr)):
        y_axis_values[i] = 0

    y_axis_values[10] = 1

    return y_axis_values

x_axis_values = np.linspace(-10, 10, 21)
#unit_sample_signal
unit_sample_value_arr = unit_sample_signal_generator(10, -10, 10)
plt.figure(figsize=(14, 8))

plt.subplot(2, 2, 1)
plt.stem(x_axis_values, unit_sample_value_arr, linefmt='-', basefmt=' ', use_line_collection=True)
plt.xlabel('n')
plt.ylabel('Amplititue')
plt.grid(True)

#unit_step_singal
unit_sample_value_arr = unit_step_signal_generator(-10, 10)
plt.subplot(2, 2, 2)
plt.stem(x_axis_values, unit_sample_value_arr, linefmt='r-', basefmt='', use_line_collection=True)
plt.xlabel('n')
plt.ylabel('Amplititue')
plt.grid(True)

#unit_ramp_signal
unit_ramp_value_arr = unit_ramp_signal_generator(-10, 10)

plt.subplot(2, 2, 3)
plt.stem(x_axis_values, unit_ramp_value_arr, linefmt='g-', basefmt='', use_line_collection=True)
plt.xlabel('n')
plt.ylabel('Amplititude')
plt.grid(True)


plt.show()

# Define the unit sample signal
```
