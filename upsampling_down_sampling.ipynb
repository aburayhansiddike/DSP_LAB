```python
import numpy as np
import matplotlib.pyplot as plt



def upsampler(original_signal, up_sample_factor):
    up_sample_value = np.zeros(len(original_signal) * up_sample_factor)
    x_range = np.arange(0, len(original_signal), 1)
    new_x = np.linspace(0, len(original_signal) - 1, len(up_sample_value)) 
    # print(len(y_range))
    # print(y_range)

    for i in range(len(up_sample_value)):
        idx = np.searchsorted(x_range, new_x[i])
        if idx == 0:
            up_sample_value[i] = original_signal[idx]
        elif idx == len(x_range):
            up_sample_value[i] = original_signal[-1]
        else:
            x1, x2 = x_range[idx - 1], x_range[idx]
            y1, y2 = original_signal[idx - 1], original_signal[idx]
            up_sample_value[i] = y1 + (new_x[i] - x1) * ((y2 - y1) / (x2 - x1))

    return new_x, up_sample_value

    


fs = 10
t = np.arange(0, 1, 1 / fs)
original_signal = np.sin(2 * np.pi * 2 * t)

plt.figure(figsize=(20, 10))
plt.subplot(3, 1, 1)
plt.plot(t, original_signal)
plt.stem(t, original_signal)


up_sample_factor = 10
new_time, up_sample_value = upsampler(original_signal, up_sample_factor)
plt.subplot(3, 1, 2)
plt.plot(new_time, up_sample_value)
plt.stem(new_time, up_sample_value)
plt.show()
```
