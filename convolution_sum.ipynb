```python
import numpy as np
import matplotlib.pyplot as plt



def convolution_sum(arr_x, arr_y):
    print(arr_x)
    print(arr_y)
    new_len = len(arr_x) + len(arr_y) - 1
    x_axis_value = np.arange(0, new_len)
    arr = np.empty(new_len)
    
    for i in range(new_len):
        arr[i] = 0
        for j in range(len(arr_y)):
            if i - j >= 0 and i - j < len(arr_x):
                arr[i] = arr[i] + (arr_x[i - j] * arr_y[j])
    return x_axis_value, arr


n = int(input("Number of elements of x(n) : "))
x = input("values of x(n) : ").split()
x = [int(value) for value in x]

m = int(input("Number of elements of h(n) : "))
h = input("Values of h(n) : ").split()
h = [int(value) for value in h]
# x = [2, 1, 0, 1]
# h = [1, 2 , 3, 1]


x_axis_value = np.arange(0, n)
arr_x = np.empty(n)
for i in range(len(x)):
    arr_x[i] = x[i]

plt.figure(figsize=(20, 10))
plt.subplot(2, 2, 1)
plt.stem(x_axis_value, arr_x, linefmt='g-', use_line_collection=True)
plt.grid(True)



arr_y = np.empty(m)
for i in range(m):
    arr_y[i] = h[i]


plt.subplot(2, 2, 2)
plt.stem(x_axis_value, arr_y, linefmt='b-', use_line_collection=True)
plt.grid(True)

x_axis_value, convo_sum = convolution_sum(arr_x, arr_y)

plt.subplot(2, 2, 3)
plt.stem(x_axis_value, convo_sum, linefmt='b-', use_line_collection=True)
plt.grid(True)
plt.show()

```
