```python
import numpy as np
import matplotlib.pyplot as plt


# Unit Step Sequence
def unit_sample(shift, n):
    shift *= -1
    step = []
    for i in n:
        if i + shift == 0:
            step.append(1)
        else:
            step.append(0)
    return step

# Main Function
n = np.arange(-5, 6, 1)
step = unit_sample(0, n)

plt.stem(n, step)
plt.xlabel('(n)')
plt.ylabel('u(n)')
plt.title('Unit Sample Sequence')
plt.show()
```

```python
def unit_step(shift, n):
    shift *= -1
    step = []
    for i in n:
        if i + shift >= 0:
            step.append(1)
        else:
            step.append(0)
    return step

# Main Function
n = np.arange(-5, 6, 1)
step = unit_step(3, n)

plt.stem(n, step)
plt.xlabel('(n)')
plt.ylabel('u(n)')
plt.title('Unit Step Sequence')
plt.show()
```

```python
def unit_ramp(shift, time):
    shift *= -1
    values = []
    power = 0
    for i in time:
        if i + shift >= 0:
            values.append(power)
            power = power + 1
        else:
            values.append(0)
    return values


time = np.arange(-10, 10, 1)
steps = unit_ramp(0, time)
print(len(steps))
print(len(time))
plt.stem(time, steps)
plt.xlabel('(n)')
plt.ylabel('a(n)')
plt.title('Unit Ramp Sequence')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def exponent_sequence(shift):
    shift *= -1
    time = np.arange(-20, 20, 0.5)
    a = 1.1
    steps = []
    for i in range(len(time)):
        if time[i] + shift >= 0:
            steps.append(a ** (time[i] + shift))
        else:
            steps.append(0)
    
    return time, steps

time, steps = exponent_sequence(0)

plt.figure(figsize=(10, 8))
plt.stem(time, steps)
plt.show()
```
